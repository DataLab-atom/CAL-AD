````python
import os
import numpy as np
from typing import Callable, List, Tuple
from reward_functions import RewardModel

# 假设 RewardModel 类已经定义，如您提供的代码所示

# 辅助函数
def initialize_population(n_pop: int, n_decap: int) -> np.ndarray:
    """初始化种群"""
    population = np.array([np.random.choice(n_decap, 10, replace=False) for _ in range(n_pop)])
    return population

def selection(population: np.ndarray, fitnesses: np.ndarray, n_selected: int) -> np.ndarray:
    """基于适应度的选择操作"""
    selected_indices = np.argsort(fitnesses)[:n_selected]
    return population[selected_indices]

def crossover(parents: np.ndarray, n_offspring: int) -> np.ndarray:
    """单点交叉操作"""
    offspring = []
    for _ in range(n_offspring):
        parent_indices = np.random.choice(len(parents), 2, replace=False)
        parent1, parent2 = parents[parent_indices[0]], parents[parent_indices[1]]
        crossover_point = np.random.randint(1, len(parent1))
        child = np.concatenate((parent1[:crossover_point], parent2[crossover_point:]))
        offspring.append(child)
    return np.array(offspring)

def mutation(individuals: np.ndarray, mutation_rate: float, n_decap: int) -> np.ndarray:
    """基于概率的变异操作"""
    for individual in individuals:
        for i in range(len(individual)):
            if np.random.rand() < mutation_rate:
                new_value = np.random.randint(0, n_decap)
                while new_value in individual:  # 确保没有重复值
                    new_value = np.random.randint(0, n_decap)
                individual[i] = new_value
    return individuals

def evaluate_population(population: np.ndarray, reward_model: RewardModel, n_inst: int) -> np.ndarray:
    """评估种群中每个个体的适应度"""
    fitnesses = np.array([reward_model(probe=np.random.randint(0, n_inst), pi=individual) for individual in population])
    return fitnesses

# 主函数
def run_ga(n_pop: int, n_iter: int, n_inst: int, elite_rate: float, n_decap: int, reward_model: RewardModel, mutation_rate: float = 0.01) -> float:
    """
    运行遗传算法进行优化。

    Args:
        n_pop (int): 种群大小。
        n_iter (int): 迭代次数。
        n_inst (int): 测试实例数。
        elite_rate (float): 精英保留率。
        n_decap (int): 去耦电容器的数量。
        reward_model (RewardModel): 用于评分的奖励模型。
        mutation_rate (float, optional): 变异率，默认为0.01。

    Returns:
        float: 最终平均奖励值。
    """
    population = initialize_population(n_pop, n_decap)
    n_elite = int(elite_rate * n_pop)
    n_offspring = n_pop - n_elite

    for generation in range(n_iter):
        fitnesses = evaluate_population(population, reward_model, n_inst)
        elites = selection(population, fitnesses, n_elite)
        offspring = crossover(elites, n_offspring)
        offspring = mutation(offspring, mutation_rate, n_decap)
        population = np.vstack((elites, offspring))

    final_fitnesses = evaluate_population(population, reward_model, n_inst)
    best_individual_index = np.argmax(final_fitnesses)
    best_individual = population[best_individual_index]
    best_reward = final_fitnesses[best_individual_index]
    print(f"Best Individual: {best_individual}, Best Reward: {best_reward}")

    return np.mean(final_fitnesses)
````
